```python
s=np.linspace(0,2*np.pi,50) #linear space
s
```

```python
print(s.size)
print(s.ndim)
print(s.shape)
```

```python
y=np.sin(s)
y
```

```python
x=np.cos(s)
x
```

```python
plt.figure() 
#plt.figure() က ရိုးရိုး run ရင်လိုတာ။
```

```python
plt.plot(x)
```

```python
plt.plot(y)
plt.title('Cosine')
```

```python
e1=rand(100)
e2=rand(100)*2
e3=rand(100)*10
e4=rand(100)*100
# print(e1,e2,e3,e4)
```

```python
#correlation matric
corrmatric=np.corrcoef([e1,e2,e3,e4])
corrmatric
```

```python
plt.imshow(corrmatric,cmap=plt.cm.YlGnBu) #cmap='YlGnBu'
plt.colorbar()
# plt.savefig('abc.jpg')
```

# Multiple Plot

```python
t=np.linspace(0,2*np.pi,50)
x=np.sin(t)
y=np.cos(t)

plt.figure(figsize=(6,4))
plt.subplot(2,1,1)
plt.plot(x)

plt.subplot(2,1,2)
plt.plot(y)

plt.show()
```

```python
from numpy.random import randint
data=randint(10000,size=(10,1000))
X=np.sum(data,axis=0)
# print(X)

plt.subplot(1,2,1)
plt.hist(X,color='r');

plt.subplot(1,2,2)
plt.hist(X,cumulative=True)
```

```python
plt.hist([X,X]);
```

```python
plt.hist(X,color='r',density=True);
```

```python
plt.plot(np.sin(t),label='sin')
plt.plot(np.cos(t),label='cos')
plt.legend()
plt.show()
```

```python
plt.plot(t,np.sin(t))
plt.xlabel('radian')
plt.ylabel('amplitude',fontsize='large') #fontsize=32
plt.title('sinX',fontsize=24)
```
